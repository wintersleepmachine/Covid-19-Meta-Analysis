```python
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in ls_cleaned: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stopwords,
                relative_scaling=0,
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

Our results suggest that high temperature and low humidity promote the formation of aerosol particles, while low temperature and high humidity promote droplet contact transmission 

We found a robust effect of mean daily temperature, which is best characterized within two regimes, below and above 25oC (Figure 1-C) emperatures higher than 25oC were associated with lower transmission rates (by 3.7% (CI: 1.9-5.4%) per additional degree) 
https://www.medrxiv.org/content/10.1101/2020.05.05.20092627v3.full.pdf

A slope of the metaregression line was significantly negative for mean air temperature (coefficient, –0.134; P = 0.019; Figure 1, upper panel), mean sea level air pressure (coefficient, –0.351; P = 0.001; Figure 1, middle panel), and mean daily maximum UV index (coefficient, –0.908; P = 0.012; Figure 1, lower panel), which indicated that COVID-19 prevalence decreased significantly as air temperature, air pressure, and UV index increased. (get figure) https://doi.org/10.1101/2020.05.09.20096321

https://www.medrxiv.org/content/10.1101/2020.03.26.20044420v1.full.pdf

Cold and dry environments are more favourable condition for virus survival 


One-degree Celsius increase in temperature and one percent increase in relative humidity reduce the R value by 0.026 (95% CI [-0.0395, -0.0125]) and 0.0076 (95% CI [-0.0108, -0.0045]), respectively.

Each additional Celsius degree decreases mortality rate by ~4%, while a 1% increase in relative humidity raises mortality rate by ~2%. Temperature is positively correlated with UV-index, for which one unit of increase results in a ~15% decrease in the mortality rate.

Researchers confirmed that respiratory infection was enhanced during unusually cold and low humidity conditions (Davis et al., 2016a, Davis et al., 2016b), indicating low humidity might also be an important risk factor for respiratory diseases. Therefore, the increase of COVID-19 mortality may also be related to the lower humidity in winter.

```python
import re

# r will store the sentences
r = []

for excerpt in df4['Excerpt'].values:
    sentences = excerpt.split('. ')
    
    for sentence in sentences:
        target = ['r0', 'reproductive', 'reproduction']
        if any(t in sentence.lower() for t in target):
            r.append(sentence)

final_r = []
for sentence in r:
    x = re.findall("\d+\.\d+", sentence)
    final_r.append(x)

def from_iterable(iterables):
    # chain.from_iterable(['ABC', 'DEF']) --> A B C D E F
    for it in iterables:
        for element in it:
            yield element
            
            
final_r = np.array([float(item) for item in list(from_iterable(final_r)) if float(item) < 10])
```

```python
sns.set_style("darkgrid")

fig, ax = plt.subplots(figsize=(8,8))


sns.distplot(final_r, bins=20)
plt.title('Reproduction Rate (R0)')
plt.ylabel('Freq')
plt.xlabel('R0')

plt.show()
```

```python
print('R0 mean: ', final_r.mean())
print('R0 median: ', final_r.min())
print('R0 max: ', final_r.max())
```
